## Esercizio 1.1

L'obiettivo di questo esercizio è testare il generatore di numeri pseudo-casuali. ### Data Blocking
In questo esercizio viene anche utilizzato il metodo del Data Blocking, che consiste nel dividere le misure in un numero arbitrario di $N$ blocchi, su ognuno dei quali vengono calcolate media e deviazione standard. Il Data Blocking viene utilizzato per calcolare la deviazione standard sui risultati dell'integrale, cercando di abbattere la correlazione che deriva dall'impossibilità di una totale casualità del generatore. <br>
Le medie parziali calcolate su ogni blocco verranno utilizzate per calcolare una media ed un errore progressivi, che ad ogni passo tengono conto di un blocco in più, permettendo anche di visualizzare la convergenza del metodo.

### Calcolo del $\chi^2$
Per verificare infine che la distribuzione sia effettivamente uniforme come ci si aspetta, è stato effettuato un test del $\chi^2$. <br>
Se la distribuzione è effettivamente uniforme ci si aspetta che ogni intervallo sia popolato da circa $M/N$ numeri. <br>
Dalla statistica del $\chi^2$ si ha:


\begin{equation}
    \chi^2 = \sum_0^N \frac{(n_i - M/N)^2}{M/N}
\end{equation}

Ci si aspetta dunque di trovare $\chi^2 \approx 100$ se la distribuzione è effettivamente uniforme. <br>
Questo test è stato ripetuto nella maniera descritta e il grafico riporta la distribuzione dei valori di $\chi^2$. Questa è stata fittata con quella teorica, la cui espressione è:

\begin{equation}
    f(x, N, k) = N \frac{x^{k/2-1}e^{-x/2}}{2^{k/2}\Gamma \left(\frac{k}{2} \right)}
\end{equation}

I parametri calcolati sono riportati, in particolare si nota che il valore del parametro risulta $k \approx 99$, in accordo con il fatto che tale parametro rappresenti il numero di gradi di libertà.

```python
from scipy.stats import chi2

# Funzione da fittare: distribuzione chi-quadro
def chi2_pdf(x, df, loc, scale):
    return chi2.pdf(x, df, loc=loc, scale=scale)

# Carico i dati
n_bins=30
chi=np.loadtxt("1.1/chiresults.out", usecols=(0), unpack='true')

#Creazione istrogramma e centri del fit
plt.figure(figsize=(9, 5))
counts, bins, _ = plt.hist(chi, n_bins, range=(60,140), density=True, edgecolor='black')     #non ho interesse in patches, assegno _
bin_centers = 0.5 * (bins[1:] + bins[:-1])

# Fit della curva
popt, pcov = curve_fit(chi2_pdf, bin_centers, counts, p0=[10, 0, 1])
df_fit, loc_fit, scale_fit = popt
print(f"Fit: dof = {df_fit:.2f}, loc = {loc_fit:.2f}, scale = {scale_fit:.2f}")

# Curva fittata
x_fit = np.linspace(60, 140, 500)
y_fit = chi2_pdf(x_fit, *popt)
plt.plot(x_fit, y_fit, 'r-', label=f'Fit $\chi^2$ (dof={df_fit:.1f})')

#Stile del grafico
plt.title("Test $\chi^2$")
plt.xlabel("$\chi^2$")
plt.ylabel("Frequenza (normalizzata)")
plt.grid(True)
plt.legend()
plt.show()

print('Chi^2 medio (atteso 100): ')
print(np.mean(chi))
print('\nVarianza (attesa 200): ')
print(np.var(chi))
print('\nErrore statistico (atteso 14.1): ')
print(np.std(chi))

```

## Esercizio 1.2: Teorema del limite centrale

In questo esercizio vogliamo estendere le funzionalità del Generatore di numeri Pseudo-Random e verificare la validità del Teorema del Limite Centrale (CLT), per il quale quando variabili casuali sono mediate tra loro durante l'estrazione, le medie tendono a distribuirsi come una gaussiana centrata sul "valore vero", indipendentemente dalla distribuzione con cui sono estratte, ammesso che questa abbia varianza finita.

In altre parole: se estraggo una variabile casuale con distribuzione ad esempio esponenziale e faccio direttamente l'istogramma delle estrazioni, ottengo un'esponenziale. Se estraggo le variabili a gruppi e ne salvo la media per ciascun gruppo, all'aumentare della dimensione del gruppo la distribuzione salvata tende a una gaussiana.

Aggiungiamo due metodi al nostro RNG: 
- Uno tale che campioni numeri secondo la distrubuzione esponenziale $p(x) = \lambda e^{-\lambda x}$, $x\in [0;+\infty]$,
- Un altro tale che campioni dalla distribuzione Lorentziana $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$.

Si ottiene questo sfruttando il metodo dell'inversione delle distribuzioni cumulative. Questo metodo si basa sul fatto che, data $x$ variabile casuale continua, se la sua funzione di ripartizione $y = F(x)$ è strettamente crescente, allora $y$ è distribuito uniformemente tra $[F(x_{min}), F(x_{max})]$. Quindi se è possibile invertire analiticamente la funzione cumulativa in $x=F^{-1}(y)$, estraendo le $y$ uniformemente è possibile ottenere delle $x$ distribuite secondo la distribuzione cercata. <br>

Sfruttiamo i nuovi metodi per fare tre istogrammi: questi verranno riempiti da $10^4$ dati del tipo $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (per $N=1, 2, 10, 100$), dato $x_i$ un numero pseudo-casuale estratto tramite un dado uniforme, esponenziale, lorentziano. Di seguito i modules utili e alcuni dati ricorrenti.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

N = [1,2,10,100]

color = ['orange', "red", "blue", "green"]
labels = ['$\mathbf{S_1}$','$\mathbf{S_2}$','$\mathbf{S_{10}}$','$\mathbf{S_{100}}$']
bins = 100
```

### Dado uniforme

```python
plt.figure(figsize=(9,6))

for i in range(4):
    #Skiprows serve per prendere ogni volta i dati a gruppi di 10^4, maxrows blocca lo scorrimento del file dopo 10^4 numeri
    unif = np.loadtxt("1.2/unif.out", skiprows = i*10000, max_rows = 10000, usecols=(0), unpack = 'true') 
    plt.hist(unif, bins, color = color[i], density='true', alpha=0.5, label = labels[i])

parameter = norm.fit(unif)      # fit gaussiano su unif
x = np.linspace(0, 1, bins)     # estremi di x, passi
y = norm.pdf(x, *parameter)     # calcola probability density function
plt.plot(x, y, color = 'black', linestyle = '-', label = "Fit Gaussiano\n $\mu$: "+str(parameter[0])+"\n $\sigma$: "+str(parameter[1]))

plt.title("Dado uniforme", fontsize =20)
plt.xlabel("x",fontsize = 10)
plt.ylabel("p(x)", fontsize = 10)
plt.grid(True)
plt.xlim(0, 1)
plt.legend()
plt.show()
```

### Dado esponenziale

```python
plt.figure(figsize=(9, 6))

for i in range(4):
    #Skiprows serve per prendere ogni volta i dati a gruppi di 10^4, maxrows blocca lo scorrimento del file dopo 10^4 numeri
    exp = np.loadtxt("1.2/exp.out", skiprows = i*10000, max_rows = 10000, usecols=(0), unpack = 'true') 
    plt.hist(exp, bins, color = color[i], density='true', alpha = 0.5, label = labels[i])

parameter = norm.fit(exp)      # fit gaussiano su unif
x = np.linspace(-0.5, 2, bins)     # estremi di x, passi
y = norm.pdf(x, *parameter)     # calcola probability density function
plt.plot(x, y, color = 'black', linestyle = '-', label = "Fit Gaussiano\n $\mu$: "+str(parameter[0])+"\n $\sigma$: "+str(parameter[1]))

plt.title("Dado Esponenziale")
plt.xlabel("x", fontsize=15)
plt.ylabel("p(x)", fontsize=15)
plt.grid(True, alpha = 0.2)
plt.xlim(0, 4)
plt.legend()
plt.show()
```

### Dado Lorentziano

```python
from scipy.stats import cauchy

plt.figure(figsize=(9,6))
for i in range(4):
    #Skiprows serve per prendere ogni volta i dati a gruppi di 10^4, maxrows blocca lo scorrimento del file dopo 10^4 numeri
    lorentz = np.loadtxt("1.2/lorentz.out", skiprows = i*10000, max_rows = 10000, usecols=(0), unpack = 'true') 
    plt.hist(lorentz, bins, range = (-11, 11), color = color[i], density='true', alpha = 0.5, label = labels[i])
    #Nota: obbligatorio chiedere un range in quanto la varianza è infinita --> Python mette gli estermi a - infinito e +infinito

parameter = cauchy.fit(lorentz)      # fit gaussiano su unif
x = np.linspace(-7, 7, bins)     # estremi di x, passi
y = cauchy.pdf(x, *parameter)     # calcola probability density function
plt.plot(x, y, color = 'black', linestyle = '-', label = "Fit Lorentziano\n $\mu$: "+str(parameter[0])+"\n $\sigma$: "+str(parameter[1]))

plt.title("Dado Lorentziano")
plt.xlabel("x", fontsize=15)
plt.ylabel("p(x)", fontsize=15)
plt.grid(True)
plt.xlim(-10, 10)
plt.legend()
plt.show()
```

#### **Osservazioni**
Nei primi due casi è evidente la convergenza alla distribuzione normale, mentre nell'ultimo istogramma la distribuzione lorenziana rimane sostanzialmente immutata: questo risultato è compatibile con il teorema del limite centrale, il quale infatti vale solo sotto l'assunzione che la varianza della distribuzione di partenza sia finita, cosa che la Lorentziana non ha.

## Esercizio 1.3
In questo ultimo esercizio si è voluto simulare l'esperimento di Buffon per il calcolo del valore di $\pi$. <br>
In questo esperimento un bastoncino di lunghezza $L=1$ viene lanciato su una griglia con linee orizzontali a distanza $d=5$. La probabilità che il bastoncino intersechi una delle linee è $P = 2L/\pi d$. Stimando la probabilità come $P \approx N_{hit}/N_{tot}$ si ottiene la seguente formula:

\begin{equation}
    \pi = \lim_{N \to \infty} \frac{2 L N_{tot}}{N_{hit} d}
\end{equation}

Considerando $d$ la distanza fra due strisce e $l$ la lunghezza del bastoncino, si estrae uniformemente un valore per $d$ nell'intervallo $ 0,d/2]$ e lo chiamiamo distance. In seguito si estrae un valore per $\theta$ nell'intervallo $[0, \pi]$ a rappresentare come è disposto angolarmente il bastoncino rispetto alla striscia. Se viene rispettata la condizione: <br>
**if (($l$ / 2.0) * sin($\theta$) >= distance)** <br>
allora possiamo dire che bastoncino e striscia si incrociano.

Infine si è usato il metodo di data blocking con 100 blocchi per valutare la convergenza di $\pi$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

```python


x, pi_sum, err_prog = np.loadtxt("1.3/piresults.out", usecols =(0,1,2), unpack = 'true')
plt.errorbar(x, pi_sum-pi, yerr=err_prog) #Funzione per mostrare l'errore 
plt.xlabel("Block Number") #Nome dei dati sull'asse X
plt.ylabel("<pi> - $\pi$") #Nome dei dati sull'asse Y
plt.title("$\pi$ convergence", fontsize=15) #titolo del grafico
plt.grid(True) #Serve per mostrare la griglia sul grafico
plt.show() #Restituisce il grafico

print("L'ultimo valore calcolato è " + str(pi_sum[-1])) 
print("Il relativo errore è " + str(err_prog[-1]))
sigma = (pi_sum[-1] - pi)/err_prog[-1]
print("Il valore si trova a " + str(round(sigma,1)) + " dal valore atteso di PI")
```
