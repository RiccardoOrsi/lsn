## Esercizio 2.1

L'obiettivo dell'esercizio è eseguire un'integrazione tramite metodi montecarlo del seguente integrale:

\begin{equation}
    I = \int_0^1 \frac{\pi}{2} \cos \left( \frac{\pi x}{2}\right) = 1
\end{equation}

Si useranno due metodi diversi fra loro per il calcolo di questo integrale:
1. Il metodo della media, con campionamento uniforme nell'intervallo $[0,1)$.
2. Il metodo dell'importance sampling, che permette di estrarre numeri casuali secondo una distribuzione di probabilità fissata (e potenzialmente intelligente): in questo modo vengono campionate più intensamente zone del dominio in cui la funzione ha valori significativamente diversi da $0$. Un limite nella ricerca della sta nel fatto che essa dev'essere **positiva** nel dominio, **finita** (eccetto in punti a misura nulla) e **normalizzata** sul dominio di integrazione.
   
La distribuzione scelta per l'importance sampling è la distribuzione lineare $p(x)=-2x+2$; questo perché per l'integrale che andremo a calcolare p(x) è gia normalizzata e definita positiva su [0, 1]. Utilizzando il meteodo dell'importance sampling ci aspettiamo che l'integrale converga più velocemente e in modo più effeiciente al risultato atteso. 

#### Librerie globali

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, ev_sum, err_prog = np.loadtxt("2.1/results_mean.out", usecols =(0,1,2), unpack = 'true')
y, ev_sum2, err_prog2= np.loadtxt("2.1/results_importance.out", usecols=(0,1,2), unpack ='true')
plt.errorbar(x, ev_sum, yerr=err_prog, ecolor="blue", label="p(x)=1") #Grafico per mostrare convergenza dell'integrale con campionamento da p(x)=1
plt.errorbar(y, ev_sum2, yerr=err_prog2, ecolor="orange", label="p(x)=-2x+2") #Grafico per mostrare convergenza dell'integrale con campionamento da p(x)=-2x+2
plt.xlabel("Block Number") #Nome dei dati sull'asse X
plt.ylabel("I value") #Nome dei dati sull'asse Y
plt.title("Integrale con due p(x) diverse", fontsize=15) #titolo del grafico
plt.grid(True) #Serve per mostrare la griglia sul grafico
plt.legend()
plt.show() #Restituisce il grafico

print("Gli ultimi valori calcolati sono " + str(ev_sum[-1]) + " e " +str(ev_sum2[-1]) + " per p(x)=1 e p(x)=-2x+2") 
print("I relativi errori sono " + str(err_prog[-1]) + " e " +str(err_prog2[-1]) )
sigma1 = (ev_sum[-1] - 1)/err_prog[-1]
sigma2 = (ev_sum2[-1] -1)/err_prog2[-1]
print("I valori si trovano a " + str(round(sigma1,1)) + "\u03C3 e "+ str(round(sigma2, 1)) + "\u03C3 dal valore atteso di 1")
```

## Esercizio 2.2

L'obiettivo di questo esercizio è realizzare due tipi di random walk (RW):

1. Su un reticolo cubico tridimensionale di lato a=1, in cui a ogni tempo discreto viene eseguito un passo
2. Nel continuo: in coordinate sferiche ($\rho, \theta, \phi$) ad ogni tempo discreto viene compiuto un "passo" di |a|=1

Si faranno due grafici tali in cui verrà mostrato l'andamento di $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ per entrambi i RW, con la loro incertezza espressa in funzione dei "passi" $i\in [0,10^2]$. Inoltre, gli verrà affiancato un fit con una curva $f(N)=k\sqrt{N}$, dove N consiste nel numero dei blocchi

#### Random Walk Discreto

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

Per simulare il random walk discreto si è utilizzata una funzione `disc_walk`, collocata in `librerie.h`. Questa funziona nel modo seguente:
- Estrae un numero casuale reale in [-1,1] e divide l'intervallo in 6 sotto-intervalli uguali
- Ogni sotto-intervallo è associato a una direzione ed ad un segno, che viene poi restituito ad un vector<double> di dimensione 3 (x-y-z)
- Da questo vector poi si calcola $r^2$

Si forniscono i grafici del fit e una rappresentazione grafica di un random walk costituito da $100$ passi, più un dato inserito manualmente per indicare l'origine

```python
#Funzione per il fit
def f (x, k):
    return k * np.sqrt(x)

fig, axes = plt.subplots(1, 2, figsize=(20, 6))    #sublpot mi permette di inserire più grafici nello stesso spazio, dividendo in righe/colonne
plt.suptitle("Random Walk Discreto", fontsize = 25)    #titolo del subplot

step, dist = np.loadtxt("2.2/results_discrete.out", usecols =(0,1), unpack = 'true')
axes[0].plot(step, dist, label='Discrete Walk', marker = '.')
axes[0].set_xlabel('step')
axes[0].set_ylabel(r'$\sqrt{\langle |\mathbf{r}|^2 \rangle}$')
axes[0].grid(True)
#Fitting function
popt, pcov = curve_fit(f, step, dist)
axes[0].plot(step,f(step,*popt), color='red', label='Fit: k='+str(*popt))
axes[0].legend()

step, dist, err = np.loadtxt("2.2/results_discrete.out", usecols =(0,1,2), unpack = 'true')
axes[1].errorbar(step, dist - f(step,*popt), yerr = err)
axes[1].set_xlabel('step')
axes[1].set_ylabel(r'$\sqrt{\langle |\mathbf{r}|^2 \rangle} - k \sqrt{N}$')
axes[1].axhline(y=0, color = 'r', linestyle = '-')
axes[1].grid(True)

plt.plot()
```
