### La teoria di Black-Scholes

La teoria di Black-Scholes assume che l'evoluzione temporale del prezzo di un'attività sia stocastica e segua un moto browniano geometrico (GBM), con tasso d’interesse privo di rischio costante $r$ e volatilità costante $\sigma$.

In finanza un'opzione è un contratto tra due (o più) parti di cui una assume un onere mentre l'altra può esercitare un diritto.
L'opzione più semplice è quella europea: chi vende l'opzione (diciamo A) garantisce a chi la compra (B) il diritto a comprare o a vendere un certo asset ad un prezzo fissato ad un istante fissato (scadenza del contratto). NOTA: solo A è vincolato dal contratto, mentre B può decidere di esercitare il diritto di vendere o comprare l'asset solo se ne ricava un profitto.

Soluzione <span style="color:blue">analitica di Black-Scholes</span> per il prezzo al tempo t<T:

- European Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- European Put-option price:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

dove
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$


dove $d_2 = d_1 - \sigma\sqrt{T-t}$, e dove
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$

è la funzione cumulativa della gaussiana $\sim \mathcal{N}(\mu=0,\sigma^2=1)$, e dove:
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$
è la funzione errore



Questa soluzione analitica ha per ipotesi due assunzioni:
- Il mercato deve essere liquido: ogni investitore deve poter comprare e vendere asset in ogni momento
- Il mercato deve essere senza frizioni: non ci devono esssere costi (tasse) legati alle transazioni

Non appena abbandoniamo le assunzioni di Black-Scholes, ad esempio quando S(t) non segue un moto browniano geometrico con deriva e volatilità costanti oppure vogliamo trattare delle opzioni "esotiche", allora non abbiamo più una soluzione analitica. Possiamo però usare metodi Monte Carlo.

Consideriamo un'attività con prezzo S(t) e un'opzione call europea ad essa collegata; supponiamo che l’evoluzione di S(t) sia data da $ S \sim \text{GBM}(r, \sigma^2)$.

- **Profitto alla scadenza** ( t = T ):

  $$ (S(T) - K)^+ := \max \left( 0, S(T) - K \right)$$

- **Valore attuale del profitto** (tempo  t ):

Per ottenere il valore attuale del profitto, dobbiamo scontarlo con un fattore $ e^{-rT} $, dato l’interesse che una banca avrebbe garantito con un deposito al tempo iniziale $ t_0 = 0 $. Infatti:

  $$
  \begin{aligned}
  t = t_0 &\Rightarrow \text{valore iniziale: } e^{-rT} (S(T) - K)^+ \\
  t = T &\Rightarrow \text{valore finale: } e^{rT} \cdot e^{-rT} (S(T) - K)^+ = (S(T) - K)^+
  \end{aligned}
  $$

- **Prezzo dell’opzione come valore atteso del profitto scontato**:

  $$\mathbb{E} \left[ e^{-rT} (S(T) - K)^+ \right] = e^{-rT} \, \mathbb{E} \left[ (S(T) - K)^+ \right]$$

- Questa quantità rappresenta la stima di quanto il detentore dell’opzione dovrebbe pagare al tempo \( t = 0 \), sulla base del profitto atteso al tempo t = T.

- Questo valore medio può essere facilmente ottenuto tramite una simulazione Monte Carlo.


 Usando i seguenti parametri: 
- Prezzo dell'asset a $t=0$: $S(0)=100$
- $T=1$
- $K=100$
- $r=0.1$
- $\sigma=0.25$


Calcolare per $t=0$ **via Monte Carlo** la plain vanilla call-option price, $C[S(0),0]$ e la put-option price, $P[S(0),0]$ con due metodi diversi:
- Metodo diretto: campionando direttamente il prezzo  finale dell'asset $S(T)$
- Metodo discretizzato: consiste nel campionare il prezzo a tempi discreti $S(t_1), S(t_2), S(t_3)...$ sfruttando il fatto che siccome gli incrementi di $W$ sono indipendenti e distribuiti secondo gaussiana possiamo usare una tecnica ricorsiva per calcolarli:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    dove $Z_i \sim \mathcal{N}(0,1)$

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 6))
plt.suptitle("Metodo Diretto", fontsize = 25)

x, sum_prog, err_prog = np.loadtxt("3.1/direct.out", usecols =(0,1,2), unpack = 'true')
axes[0].errorbar(x, sum_prog-14.97579, yerr = err_prog)
axes[0].set_xlabel('Block #')
axes[0].set_ylabel(r'$C[S(0),0]$')
axes[0].set_title("Call Option", fontsize=15)
axes[0].grid(True)
axes[0].axhline(y=0, color = 'r', linestyle = '-')

print("\033[1m" + "Call Option" + "\033[0m")
print("L'ultimo valore calcolato è " + str(sum_prog[-1])) 
print("Il relativo errore è " + str(err_prog[-1]))
sigma = (sum_prog[-1] - 14.97579)/err_prog[-1]
print("Il valore si trova a " + str(round(sigma,1)) + "\u03C3 dal valore atteso di 14.97579")

x, sum_prog, err_prog = np.loadtxt("3.1/direct.out", usecols =(0,3,4), unpack = 'true')
axes[1].errorbar(x, sum_prog-5.4595, yerr = err_prog)
axes[1].set_xlabel('Block #')
axes[1].set_ylabel(r'$P[S(0),0]$')
axes[1].set_title("Pull Option", fontsize=15)
axes[1].grid(True)
axes[1].axhline(y=0, color = 'r', linestyle = '-')
plt.plot()

print("\033[1m" + "Pull Option" + "\033[0m")
print("L'ultimo valore calcolato è " + str(sum_prog[-1])) 
print("Il relativo errore è " + str(err_prog[-1]))
sigma = (sum_prog[-1] - 5.4595)/err_prog[-1]
print("Il valore si trova a " + str(round(sigma,1)) + "\u03C3 dal valore atteso di 5.4595")

plt.plot()
```
