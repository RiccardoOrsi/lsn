# ESERCITAZIONE 4

### INTRODUZIONE

Questa esercitazione verge sull'utilizzo di una serie di librerie utili allo svolgimento di simulazioni di dinamica molecolare, le quali sono contenute nella cartella NSL_SIMULATOR. L'input da noi fornito consisterà nelle proprietà microscopiche del sistema che vogliamo simulare e, in base alla sua evoluzione, vogliamo ricavarne le proprietà macroscopiche. Nello specifico le proprietà di nostro interesse sono:

- $E/N$: Energia totale per numero di particelle
- $U/N$: Energia potenziale per numero di particelle
- $K/N$: Energia cinetica per numero di particelle
- $T$: Temperatura
- $P$: Pressione

Tutte queste simulazioni verranno eseguite secondo un potenziale di Lennard-Jones:

\begin{equation}
    V_{LJ}(r) = 4 \varepsilon \left[  \left(\frac{\sigma}{r}\right)^{12} - \left( \frac{\sigma}{r}\right)^6\right]
\end{equation}

il quale verrà espresso in unità ridotte, utili per confronti tra materiali diversi e per identificare due sistemi all'apparenza diversi, ma una volta tradotti in unità ridotte di fatto identici. <br>
Un'altra assunzione da noi fatta durante queste simulazioni sarà di trovarci nell'<b>ensemble microcanonico</b>. Poiché non possiamo conoscere lo stato microscopico esatto, adottiamo una descrizione probabilistica: immaginiamo un ensemble di sistemi tutti nello stesso stato macroscopico ma con microstati diversi. Nello specifico l'ensemble microcanonico è tale per cui tutti gli stati microscopici compatibili con un’energia $E$ hanno uguale probabilità. Il sistema considerato deve quindi essere isolato, di conseguenza rimarranno constanti anche il numero di particelle $N$ e il volume $V$

### APPROSSIMAZIONI

Sistemi composti da un gran numero di particelle, oltre a essere analiticamente insolubili, diventano molto presto anche inefficienti da un punto di vista computazionale. Questo perchè mantenere in memoria le caratteristiche di un gran numero di particelle porta molto presto il tempo di esecuzione a divergere. Sono quindi necessarie alcune approssimazioni e semplificazioni al fine di alleggerire il computer nell'esecuzione del programma. Introduciamo quindi:

- <b>Periodic boundary condition (PBC)</b>: simulazioni con poche particelle ($N<10^5$) soffrono di effetti di superficie in quanto le particelle ai bordi subiscono forze molto diverse da quelle nel bulk. Le PBC quindi replicano il sistema in tutte le direzioni per simulare un sistema “infinito": si replica la scatola simulata (cubo) in tutte le direzioni dello spazio. In più, una particella esce da un lato della scatola, entra da quello opposto, come se lo spazio fosse “toroidale”. Adottiamo inoltre la <b>Minimum image convention</b>: ogni particella interagisce solo con l’immagine più vicina delle altre.
- <b>Raggio di Cutoff</b>: Il potenziale di Lennard-Jones ha $r\in (0, \infty)$, quindi teoricamente tutte le particelle interagiscono fra di loro indipendentemente da $r$, ma per la sua forma i contributi maggiori sono quelli a corto raggio. Per potenziali a corto raggio, si impone un cut-off: $v(r)=0$ per $r>r_{c}$. La distanza di cut-off dev’essere $r_{c}<L/2$ per coerenza con la minimum image convention. Si applicano però correzioni a energia e pressione per "compensare" i fattori esclusi dal cut-off.

### ALGORITMO RISOLUTIVO

Si utilizzerà l'algoritmo di Verlet: un metodo di integrazione delle equazioni di Newton per simulare il moto delle particelle. Permette di aggiornare le posizioni delle particelle nel tempo, conoscendo la loro accelerazione (cioè la forza) e la posizione attuale e precedente.
\begin{equation}
    \vec{r}(t + \delta t) \approx 2 \vec{r} (t) - \vec{r}(t - \delta t) + \delta t^2 \vec{a} (t)
\end{equation}
Notare che siccome servono sia $r(t+\delta t)$ sia $r(t-\delta t)$ non possiamo limitarci alle condizioni inziali $r(0)$ e $v(0)$ per avviare la simulazione. In compenso, però l'algoritmo di Verlet conserva molto bene l'energia sul lungo periodo, è reversibile temporalmente, è semplice da implementare ed è computazionalmente efficiente

## ESERCIZIO 4.1

Il primo esercizio consiste nell'implementazione del calcolo della pressione tramite la seguente formula, ricavabile dal teorema del viriale:
\begin{equation}
    P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j > i}^{N} 48 \varepsilon \left[ \left( \frac{\sigma}{| \vec{r}_i - \vec{r}_j |} \right)^{12} - \frac{1}{2}\left( \frac{\sigma}{| \vec{r}_i - \vec{r}_j |} \right)^6 \right]\right\rangle
\end{equation}

Esprimendo le grandezze in unità di Lennard Jones e notando che l'espressione tra le parentesi angolari corrisponde al viriale si può semplificare l'espressione:

\begin{equation}
    P^* = \rho^* T^* - \frac{\langle W \rangle}{3V^*}
\end{equation}

Si è tenuto inoltre conto del fatto che le sommatorie non debbano considerare tutte le particelle ma solo quelle con $r< r_{c}$.


## ESEERCIZIO 4.2

La consegna richiede di eseguire tre simulazioni di dinamica molecolare per l'Argon $\sigma = 0.34$nm, $\epsilon/k_B = 120$K, $m=39.948$amu nelle seguenti condizioni:

1. Stato solido: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio cut-off: $r_c = 2.2$)
2. Stato liquido: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio cut-off: $r_c = 2.5$)
3. Stato gassoso: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio cut-off: $r_c = 5.0$)

Mostrare attraverso grafici i valori ottenuti per le medie successive e le relative incertezze, in funzione del numero dei blocchi, delle seguenti grandezze:

- Energia per particella, $E/N$
- Energia potenziale per particella, $U/N$
- Energia cinetica per particella, $K/N$
- Temperatura, $T$
- Pressione, $P$

Nello svolgimento delle simulazioni però si nota come la Temperatura nel tempo tenda a divergere da quella data in input. Questo perché noi per comodità poniamo le particelle inizialmente in una struttura cristallina per comodità, ma con la simulazione questa si perde in quanto il sistema (che inizialmente si può assumere come allo stato solido) non è stabile da un punto di vista termodinamico alla data temperatura $T$. Questo va contro l'interesse della simulazione in quanto le quantità calcolate sono richieste all'equilibrio. Bisogna quindi trovare il valore di input $T_{0}$ tale che all'equilibrio $T_{eq}=T^\star$



#### TERMALIZZAZIONE

Sopra sono riportate le librerie comuni, di seguito invece le simulazioni svolte per iungere alla termalizzazione. Per vedere i risultati sul lungo periodo sono state fatte simulazioni da 20000 blocchi, ma con step $n=1$

```python
import matplotlib.pyplot as plt
import numpy as np

Teq = [1.2, 1.1, 0.8]
Tzero = ['$T_0 = 0.95$', '$T_0 = 2$', '$T_0 = 1.55$']
rep = ['tgas', 'tliquid', 'tsolid']
state = ["Gas", "Liquido", "Solido"]

for i in range(0,3):
    x, aveT = np.loadtxt("NSL_SIMULATOR/OUTPUT/"+rep[i]+"/temperature.dat", skiprows=1, usecols=(0, 1), delimiter=None, unpack='True')
    #Uso delimiter=None in modo che possa fare autodetect delle spaziature
    fig, ax = plt.subplots(figsize=(15, 3))
    ax.plot(x, aveT, marker=None, label="Temperatura")
    ax.set_title("Termalizzazione " + state[i] + " in " + Tzero[i], fontsize=18)
    ax.set_ylabel(r'$T_{eq}$', fontsize=15)
    ax.set_xlabel('#block', fontsize=15)
    ax.axhline(y=Teq[i], color="r", label="Temperatura di equilibrio: " + str(Teq[i]))
    ax.grid(True)
    ax.legend()
    plt.show()
```

Possiamo notare come Per lo stato solido e liquido si raggiunga la termalizzazione in relativamente pochi blocchi, mentre per il gas, la si raggiunge dopo molti più blocchi; ciò avviene a causa della densità molto piu bassa del gas rispetto agli altri due stati della materia, riducendo quindi le interazioni tra le varie coppie di particelle.

Possiamo ora effettuare la simulazione del sistema nei tre stati della materia, avendo trovato le tre $T_0$ necessarie:

### STATO SOLIDO

```python
import matplotlib.pyplot as plt
import numpy as np

quantity=["kinetic_energy", "potential_energy", "pressure", "temperature", "total_energy"]
color=["yellow", "orange", "red", "purple", "blue"]

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
plt.suptitle("STATO SOLIDO")

for i in range(5):
    x, ave, error = np.loadtxt("NSL_SIMULATOR/OUTPUT/solid/"+quantity[i]+".dat", skiprows=1, usecols=(0, 2, 3), delimiter=None, unpack='True')
    ax=axes[i//3, i%3] #array bidiimensionale, faccio scorrere la riga tramite il calcolo della divisione per interi, le colonne col calcolo del resto
    ax.errorbar(x, ave, yerr=error, marker="o", color=color[i])
    ax.set_title(quantity[i], fontsize=20)
    ax.set_ylabel(quantity[i], fontsize=15)
    ax.set_xlabel('#block', fontsize=15)
    ax.grid(True)

axes[1,2].remove() #altrimenti ci sarebbe un grafico nella seconda riga, terza colonna
fig.tight_layout()
plt.show()
```
